## DFS를 이용해 모든 음식점을 방문하는 에니메이션 코드

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import networkx as nx
import platform

# 1. 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Windows': plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin': plt.rcParams['font.family'] = 'AppleGothic'
else: plt.rcParams['font.family'] = 'NanumGothic'

# 2. 그래프 데이터 정의
nodes = {
    0: "오몬자", 1: "남산광어", 2: "와이인", 3: "효뜨", 4: "문화식당",
    5: "따로집", 6: "스페이스", 7: "코지밀", 8: "르부아", 9: "소테"
}

edges = [
    (0, 1, 37), (0, 2, 47),
    (1, 2, 29),
    (2, 3, 21),
    (3, 4, 36), (3, 5, 19),
    (4, 5, 48),
    (4, 6, 22),
    (5, 7, 42),
    (6, 8, 18),
    (7, 9, 2)
]

G = nx.Graph()
for node_id, node_name in nodes.items(): G.add_node(node_id, name=node_name)
for u, v, w in edges: G.add_edge(u, v, weight=w)

# 시각화 배치 수정
pos = {
    0: (0, 4),
    1: (-1.2, 3), 2: (1.2, 3), # 1-2 간선이 보이게 벌림
    3: (0, 2),
    4: (-1.2, 1), 5: (1.2, 1), # 4-5 간선이 보이게 벌림
    6: (-0.6, 0), 7: (1.2, 0),
    8: (-0.6, -1), 9: (1.2, -1)
}

# 3. DFS 경로 및 프레임 생성
visited_nodes, visited_edges, frames = [], [], []
current_nodes, current_edges = set(), set()

def dfs(node, parent=None):
    if node in visited_nodes: return
    visited_nodes.append(node)
    current_nodes.add(node)
    if parent is not None:
        visited_edges.append(tuple(sorted((parent, node))))
        current_edges.add(tuple(sorted((parent, node))))
    frames.append((list(current_nodes), list(current_edges)))

    # 인접 노드 번호순 방문
    for neighbor in sorted(list(G.neighbors(node))):
        if neighbor not in visited_nodes:
            dfs(neighbor, node)

dfs(0) # 0번 오몬자 시작
for _ in range(3): frames.append((list(current_nodes), list(current_edges))) # 완료 프레임 추가

# 4. 애니메이션 설정
fig, ax = plt.subplots(figsize=(7, 9))
node_labels = {k: f"{k}\n{v}" for k, v in nodes.items()}
edge_labels = {(u, v): f"{d['weight']}" for u, v, d in G.edges(data=True)}

def update(frame):
    ax.clear()
    ax.set_title("음식점 방문 과정 (DFS - 수정된 그래프)", fontsize=13, fontweight='bold', pad=20)
    act_nodes, act_edges = frame
    node_colors = ['#FFD54F' if node in act_nodes else '#E0E0E0' for node in G.nodes()]
    edge_colors = ['#FF5722' if tuple(sorted((u, v))) in act_edges else '#BDBDBD' for u, v in G.edges()]
    edge_widths = [4.0 if tuple(sorted((u, v))) in act_edges else 1.5 for u, v in G.edges()]
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1500, edgecolors='black', linewidths=1.5, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color=edge_colors, width=edge_widths, ax=ax)
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_family=plt.rcParams['font.family'], font_size=10, font_weight='bold', ax=ax)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=11, font_color='blue', ax=ax)
    ax.axis('off')

ani = animation.FuncAnimation(fig, update, frames=frames, repeat=True, interval=1200)
plt.show()
```

## BFS를 이용해 모든 음식점을 방문하는 에니메이션 코드

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import networkx as nx
import platform
from collections import deque

# 1. 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Windows': plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin': plt.rcParams['font.family'] = 'AppleGothic'
else: plt.rcParams['font.family'] = 'NanumGothic'

# 2. 그래프 데이터 정의
nodes = { 0: "오몬자", 1: "남산광어", 2: "와이인", 3: "효뜨", 4: "문화식당", 5: "따로집", 6: "스페이스", 7: "코지밀", 8: "르부아", 9: "소테" }
edges = [
    (0, 1, 37), (0, 2, 47),
    (1, 2, 29),
    (2, 3, 21),
    (3, 4, 36), (3, 5, 19),
    (4, 5, 48),
    (4, 6, 22), (5, 7, 42), (6, 8, 18), (7, 9, 2)
]
G = nx.Graph()
for node_id, node_name in nodes.items(): G.add_node(node_id, name=node_name)
for u, v, w in edges: G.add_edge(u, v, weight=w)

# 시각화 배치
pos = { 0: (0, 4), 1: (-1.2, 3), 2: (1.2, 3), 3: (0, 2), 4: (-1.2, 1), 5: (1.2, 1), 6: (-0.6, 0), 7: (1.2, 0), 8: (-0.6, -1), 9: (1.2, -1) }

# 3. BFS 경로 및 프레임 생성 (DFS와 동일한 타이밍으로 수정)
frames = []
current_nodes = set([0])
current_edges = set()

# 시작 상태 저장 (0번만 켜진 상태)
frames.append((list(current_nodes), list(current_edges)))

def bfs():
    visited = {0}
    queue = deque([0])

    while queue:
        curr = queue.popleft()

        # 번호 순서대로 인접 노드 탐색
        for neighbor in sorted(list(G.neighbors(curr))):
            if neighbor not in visited:
                visited.add(neighbor)

                # [수정] 정점과 간선을 "동시에" 추가하고 바로 프레임에 기록합니다.
                current_nodes.add(neighbor)
                current_edges.add(tuple(sorted((curr, neighbor))))
                frames.append((list(current_nodes), list(current_edges)))

                queue.append(neighbor)

bfs()

# 녹화 여유를 위해 완료 상태 프레임 추가
for _ in range(3): frames.append((list(current_nodes), list(current_edges)))

# 4. 애니메이션 설정
fig, ax = plt.subplots(figsize=(7, 9))
node_labels = {k: f"{k}\n{v}" for k, v in nodes.items()}
edge_labels = {(u, v): f"{d['weight']}" for u, v, d in G.edges(data=True)}

def update(frame):
    ax.clear()
    ax.set_title("음식점 방문 과정 (BFS)", fontsize=13, fontweight='bold', pad=20)
    act_nodes, act_edges = frame
    node_colors = ['#81C784' if node in act_nodes else '#E0E0E0' for node in G.nodes()]
    edge_colors = ['#2E7D32' if tuple(sorted((u, v))) in act_edges else '#BDBDBD' for u, v in G.edges()]
    edge_widths = [4.0 if tuple(sorted((u, v))) in act_edges else 1.5 for u, v in G.edges()]

    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1500, edgecolors='black', linewidths=1.5, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color=edge_colors, width=edge_widths, ax=ax)
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_family=plt.rcParams['font.family'], font_size=10, font_weight='bold', ax=ax)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=11, font_color='blue', ax=ax)
    ax.axis('off')

ani = animation.FuncAnimation(fig, update, frames=frames, repeat=True, interval=1200)
plt.show()
```

## 모든 음식점을 최소 시간으로 방문할 수 있는 그래프를 프림 알고리즘 코드

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import networkx as nx
import platform
import heapq

# 1. 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Windows': plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin': plt.rcParams['font.family'] = 'AppleGothic'
else: plt.rcParams['font.family'] = 'NanumGothic'

# 2. 그래프 데이터 정의
nodes = { 0: "오몬자", 1: "남산광어", 2: "와이인", 3: "효뜨", 4: "문화식당", 5: "따로집", 6: "스페이스", 7: "코지밀", 8: "르부아", 9: "소테" }
edges = [
    (0, 1, 37), (0, 2, 47),
    (1, 2, 29),
    (2, 3, 21),
    (3, 4, 36), (3, 5, 19),
    (4, 5, 48),
    (4, 6, 22), (5, 7, 42), (6, 8, 18), (7, 9, 2)
]
G = nx.Graph()
for node_id, node_name in nodes.items(): G.add_node(node_id, name=node_name)
for u, v, w in edges: G.add_edge(u, v, weight=w)

# 시각화 배치 수정
pos = { 0: (0, 4), 1: (-1.2, 3), 2: (1.2, 3), 3: (0, 2), 4: (-1.2, 1), 5: (1.2, 1), 6: (-0.6, 0), 7: (1.2, 0), 8: (-0.6, -1), 9: (1.2, -1) }

# 3. Prim 경로 및 프레임 생성
visited_nodes, frames = [0], [([0], [])]
mst_nodes, visited_edges = {0}, set()

def prim(start_node):
    candidate_edges = []
    for neighbor in G.neighbors(start_node):
        heapq.heappush(candidate_edges, (G[start_node][neighbor]['weight'], start_node, neighbor))

    while candidate_edges:
        weight, u, v = heapq.heappop(candidate_edges)
        if v in mst_nodes: continue
        mst_nodes.add(v)
        visited_edges.add(tuple(sorted((u, v))))
        frames.append((list(mst_nodes), list(visited_edges)))

        for next_neighbor in G.neighbors(v):
            if next_neighbor not in mst_nodes:
                heapq.heappush(candidate_edges, (G[v][next_neighbor]['weight'], v, next_neighbor))

prim(0) # 0번 오몬자 시작
for _ in range(3): frames.append((list(mst_nodes), list(visited_edges)))

# 4. 애니메이션 설정
fig, ax = plt.subplots(figsize=(7, 9))
node_labels = {k: f"{k}\n{v}" for k, v in nodes.items()}
edge_labels = {(u, v): f"{d['weight']}" for u, v, d in G.edges(data=True)}

def update(frame):
    ax.clear()
    # MST 결과 설명 추가
    ax.set_title("최소 시간 방문 경로 (프림 알고리즘)",
                 fontsize=12, fontweight='bold', pad=25, color='#01579B')
    act_nodes, act_edges = frame
    node_colors = ['#29B6F6' if node in act_nodes else '#E0E0E0' for node in G.nodes()]
    edge_colors = ['#0288D1' if tuple(sorted((u, v))) in act_edges else '#E0E0E0' for u, v in G.edges()] # 안쓰는간선 연하게
    edge_widths = [4.5 if tuple(sorted((u, v))) in act_edges else 1.0 for u, v in G.edges()]
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1500, edgecolors='black', linewidths=1.5, ax=ax)
    # Prim 알고리즘의 MST를 더 명확히 보여주기 위해 연한 기본 간선 위에 MST만 진하게 그림
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='#EEEEEE', width=1.0, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=act_edges, edge_color='#0288D1', width=4.5, ax=ax)
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_family=plt.rcParams['font.family'], font_size=10, font_weight='bold', ax=ax)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=11, font_color='blue', ax=ax)
    ax.axis('off')

ani = animation.FuncAnimation(fig, update, frames=frames, repeat=True, interval=1200)
plt.show()
```

## 모든 음식점을 최소 시간으로 방문할 수 있는 그래프를 크루스칼 알고리즘

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import networkx as nx
import platform

# 1. 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Windows': plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin': plt.rcParams['font.family'] = 'AppleGothic'
else: plt.rcParams['font.family'] = 'NanumGothic'

# 2. 그래프 데이터 정의
nodes = { 0: "오몬자", 1: "남산광어", 2: "와이인", 3: "효뜨", 4: "문화식당", 5: "따로집", 6: "스페이스", 7: "코지밀", 8: "르부아", 9: "소테" }
edges = [
    (0, 1, 37), (0, 2, 47),
    (1, 2, 29),
    (2, 3, 21),
    (3, 4, 36), (3, 5, 19),
    (4, 5, 48),
    (4, 6, 22), (5, 7, 42), (6, 8, 18), (7, 9, 2)
]
G = nx.Graph()
for node_id, node_name in nodes.items(): G.add_node(node_id, name=node_name)
for u, v, w in edges: G.add_edge(u, v, weight=w)

# 시각화 배치
pos = { 0: (0, 4), 1: (-1.2, 3), 2: (1.2, 3), 3: (0, 2), 4: (-1.2, 1), 5: (1.2, 1), 6: (-0.6, 0), 7: (1.2, 0), 8: (-0.6, -1), 9: (1.2, -1) }

# 3. 크루스칼 알고리즘 구현을 위한 Disjoint Set (Union-Find) 설정
parent = {i: i for i in nodes}

def find(i):
    if parent[i] == i:
        return i
    parent[i] = find(parent[i])
    return parent[i]

def union(i, j):
    root_i = find(i)
    root_j = find(j)
    if root_i != root_j:
        parent[root_i] = root_j
        return True
    return False

# 가중치 순으로 모든 간선 정렬
sorted_edges = sorted(edges, key=lambda item: item[2])

# 크루스칼 애니메이션 프레임 생성
frames = []
current_nodes = set()
current_edges = set()

# 아무것도 선택되지 않은 초기 상태 추가
frames.append((list(current_nodes), list(current_edges)))

for u, v, w in sorted_edges:
    # 사이클을 만들지 않는 경우에만 간선과 해당 노드들을 동시에 선택
    if union(u, v):
        current_nodes.add(u)
        current_nodes.add(v)
        current_edges.add(tuple(sorted((u, v))))
        # 정점과 간선이 동시에 탁 켜지도록 프레임 기록
        frames.append((list(current_nodes), list(current_edges)))

# 녹화 여유를 위해 완료 상태 프레임 추가
for _ in range(3): frames.append((list(current_nodes), list(current_edges)))

# 4. 애니메이션 설정
fig, ax = plt.subplots(figsize=(7, 9))
node_labels = {k: f"{k}\n{v}" for k, v in nodes.items()}
edge_labels = {(u, v): f"{d['weight']}" for u, v, d in G.edges(data=True)}

def update(frame):
    ax.clear()
    ax.set_title("최소 시간 방문 경로 (Kruskal's Algorithm)",
                 fontsize=12, fontweight='bold', pad=25, color='#4A148C')
    act_nodes, act_edges = frame

    # 크루스칼용 색상 (고급스러운 보라색 계열)
    node_colors = ['#BA68C8' if node in act_nodes else '#E0E0E0' for node in G.nodes()]

    # 전체 구조를 보여주기 위해 배경에 연한 기본 간선을 깔고, 그 위에 선택된 MST 간선을 진하게 덮어씌움
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1500, edgecolors='black', linewidths=1.5, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='#EEEEEE', width=1.0, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=act_edges, edge_color='#7B1FA2', width=4.5, ax=ax)
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_family=plt.rcParams['font.family'], font_size=10, font_weight='bold', ax=ax)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=11, font_color='blue', ax=ax)
    ax.axis('off')

# 화면 상에서만 1.2초 간격으로 무한 반복 재생
ani = animation.FuncAnimation(fig, update, frames=frames, repeat=True, interval=1200)
plt.show()
```
